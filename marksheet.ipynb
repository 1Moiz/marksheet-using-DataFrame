# Marksheet Using DataFrame

```python
import pandas as pd
```

```python
ms = {
    'python1' : [67,54,76,71,54,65,87,98,78],
    'python2' : [65,67,87,98,65,34,66,80,76],
    'python3' : [56,65,87,98,56,45,76,87,98],

}
ms
```

```python
marks =pd.DataFrame(ms)
```

```python
marks
```

```python
marks['Total'] = marks.python1+marks.python2+marks.python3
```

```python
marks
```

```python
marks['Percenatge'] = marks.Total/300*100
```

```python
marks
```

```python
marks['Status'] = ['Pass' if all(i > 70) else 'Fail' for i in marks[['python1','python2','python3']].values]
```

```python
marks
```

```python
marks['Names'] = ['Ali','Anwar','Asif','Akbar','Mehwish','Ushba','Alina','Sania','Ahmed']
```

```python
marks
```

```python
marks = pd.DataFrame(marks, columns=['Names','python1','python2','python3','Total','Percenatge','Status'])
```
